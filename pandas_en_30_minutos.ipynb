# Algoritmo para calcular

```python
## Importamos la libreria random para este ejemplo
import random
```

```python
## Ver la tabla hasta el momento
df = pd.read_csv('dataset.csv', index_col = "id")
df.head()
```

```python
## def es definir una función: def calcularGanancias donde se va a trabajar con la columna retweets,
## se crea una variable ganancia. 
## Para  usarla con la columna retweets donde se va usar un random aleatorio entre 3 y 5 (multiplica retweets x 3 o por 5)
## Para que retorne una nueva columna: ganancias

def calcularGanancias(retweets):
    ganancia = retweets * random.randint(3, 5)
    return ganancia

df["ganancias"] = df["retweets"].apply(calcularGanancias)
df.head(10)
```

```python
## Se crea otra función:  popularidad para dividir en una variable: resultado (follovees entre follovers)
## y se crea una nueva columna: popularidad

def calcularGanancias(retweets):
    ganancia = retweets * random.randint(3, 5)
    return ganancia

df["ganancias"] = df["retweets"].apply(calcularGanancias)

def popularidad(fila):
    resultado = fila["followees"]/fila["followers"]
    return resultado

df["popularidad"] = df.apply(popularidad, axis=1)
df.head()
```

# Agrupación de Datos

```python
## Ver la tabla hasta el momento
df = pd.read_csv('dataset.csv', index_col = "id")
df
```

```python
## Agrupando por country (pais) y sacando el promedio con mean
df.groupby("country").mean()
```

```python
## la función Agg permite agregar varias funciones. Agrupa por country pero realiza diferentes operaciones para cada columna
df.groupby("country").agg({
 "followers": 'sum',
 "mentions": 'mean',   
 "retweets": 'max'   
})
```

```python
## grouped variable para conseguir mayores a 5000 en follovers
grouped = df.groupby("country").agg({
 "followers": 'sum',
 "mentions": 'mean',   
 "retweets": 'max'   
})

grouped[grouped["followers"]>5000]
```
